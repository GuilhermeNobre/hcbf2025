```python
from ultralytics import YOLO
import numpy as np
import cv2
```

```python
model = YOLO("bacteria-yolo11n-cls.pt")
```

```python
result = model("test/ac.isr.jpeg")
```

```python
names_dict = result[0].names
names_dict
```

```python
probs = result[0].probs.data.tolist()
probs
```

```python
print(names_dict[np.argmax(probs)])
```
